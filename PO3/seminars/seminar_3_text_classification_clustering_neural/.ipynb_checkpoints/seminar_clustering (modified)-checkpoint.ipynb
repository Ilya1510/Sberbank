```python
data_to_cluster = data.copy()
```

```python
data_to_cluster['Нет задолженности по кредитам'] = data_to_cluster['Суммарная задолженность по всем кредитам'] == 0
```

```python
data_to_cluster['Суммарная задолженность по всем кредитам'] = np.log(1+data_to_cluster['Суммарная задолженность по всем кредитам'])
data_to_cluster['Общая сумма списаний со всех счетов клиента за месяц'] =\
                np.log(1+data_to_cluster['Общая сумма списаний со всех счетов клиента за месяц'])
```

```python
data_to_cluster.hist(figsize=(15,8))
plt.show()
```

```python
from sklearn.preprocessing import scale
```

```python
data_to_cluster = pd.DataFrame(scale(data_to_cluster, axis=0), columns=data_to_cluster.columns)
```

```python
data_to_cluster.describe()
```

```python
import seaborn as sns
```

```python
sns.heatmap(data_to_cluster.corr())
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=9, random_state=0).fit(data_to_cluster)
```

```python
data['cluster'] = kmeans.labels_
```

```python
means = np.stack([data[data['cluster'] == i].describe().values[1] for i in range(9)])
```

```python
pd.options.display.max_columns = 30
```

```python
means = pd.DataFrame(means, columns=data.columns)
means
```
